```python
from gensim import corpora, models
import pyLDAvis
import pyLDAvis.gensim_models
import pandas as pd
import json
```

```python
NUM_TOPICS = 3
NUM_TOPIC_WORDS=30
```

```python
def build_doc_term_mat(documents):
    dictionary=corpora.Dictionary(documents)
    corpus = [dictionary.doc2bow(document) for document in documents]
    return corpus, dictionary
```

```python
pyLDAvis.enable_notebook()
```

```python
df = pd.read_csv('data/lda_analysis_data.csv')
```

```python
df.sample(5)
```

```python
import ast
x = ast.literal_eval(x)
```

```python
df['nouns'] = df['nouns'].apply(lambda x: ast.literal_eval(x))
```

# lda 의 토픽의 번호는 바뀔수 있음

### 일베 토픽
#### 1: 야권 대선 후보, 정치인들 여론관련
#### 2: 당대표당선 관련 여론 등
#### 3:  그냥 욕하는듯?

```python
data_word = df[df['source']=='ilbe']['nouns'].tolist()

corpus, dictionary = build_doc_term_mat(data_word)
model= models.ldamodel.LdaModel(corpus, num_topics=NUM_TOPICS,
                                id2word=dictionary, 
                                alpha='auto', eta='auto')
data = pyLDAvis.gensim_models.prepare(model, corpus, dictionary)
data
```
